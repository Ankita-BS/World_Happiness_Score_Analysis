### Importing Libraries

```python
# Importing Libraries
import pandas as pd
import numpy as np
import os
import glob
```

```python
# Define Path
path = r'C:\Users\ankit\Desktop\DataAnalysisCF\DataImmersion\Advanced Analytics & Dashboard Design\World Happiness Analysis'
```

```python
# Importing all the datafarmes into 'datasets' data_dictionary
file_paths = glob.glob(os.path.join(path,'02 Data','Original Data','*.csv'))
datasets = {}
for file_path in file_paths:
    file_name = file_path.split("/")[-1].replace(".csv", "")
    datasets[file_name] = pd.read_csv(file_path)
```

```python
type(datasets)
```

```python
# Adding new column 'year' to all datasets
year = 2015
for value in datasets.values():    
    value["year"] = year
    year +=1
    
```

```python
# Converting Data_dictionary into list of data frames
df_list = [pd.DataFrame(data_dict) for data_dict in datasets.values()]
for df in df_list:
    print(df, "\n")
```

```python
# Concatinating Dataframes
combined_df = pd.concat(df_list, ignore_index=True)
combined_df
```

```python
# Exporting Concatinated Dataframe
combined_df.to_csv(os.path.join(path,'02 Data','Prepared Data','concatenated_data.csv'))
```

### Data Wrangling and Consistency Checks

```python
combined_df['country'].value_counts()
```

```python
combined_df['region'].value_counts()
```

```python
# Checking mixed type columns
for col in combined_df.columns.tolist():
  weird = (combined_df[[col]].map(type) != combined_df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (combined_df[weird]) > 0:
    print (col)
```

```python
#Checking for missing values in columns
combined_df.isnull().sum()
```

```python
# mean value of healthy_life_expectancy
hle_mean = combined_df['healthy_life_expectancy'].mean()
hle = round(hle_mean, 3)
```

```python
# Imputing mean for missing values
combined_df['healthy_life_expectancy'].fillna(hle,inplace = True)
```

```python
# mean value of perceptions_of_corruption
pc_mean = combined_df['perceptions_of_corruption'].mean()
pc = round(pc_mean, 3)
```

```python
# Imputing mean for missing values
combined_df['perceptions_of_corruption'].fillna(pc,inplace = True)
```

```python
combined_df.isnull().sum()
```

```python
#Checking duplicate entries
dups = combined_df[combined_df.duplicated()]
dups
```

```python
#Exporting cleaned dataframe
combined_df.to_csv(os.path.join(path,'02 Data','Prepared Data','wh_clean.csv'))
```
